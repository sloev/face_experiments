```python
from imutils import face_utils
import imutils
import numpy as np
import collections
import dlib
import cv2
from matplotlib import pyplot as plt
```

```python
detector = dlib.get_frontal_face_detector()
predictor = dlib.shape_predictor('landmarks.dat')
```

```python
image = cv2.imread("processed/001.jpg")
plt.imshow(image)
```

```python
def add_faces(image, background):
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    rects = detector(gray, 1)
    for rect in rects:
        shape = predictor(gray, rect)
        shape = face_utils.shape_to_np(shape)

        #initialize mask array
        remapped_shape = np.zeros_like(shape) 
        feature_mask = np.zeros((image.shape[0], image.shape[1]))   

        # we extract the face
        remapped_shape = cv2.convexHull(shape)
        cv2.fillConvexPoly(feature_mask, remapped_shape[0:27], 1)
        feature_mask = feature_mask.astype(np.bool)
        background[feature_mask] = image[feature_mask]
```

```python
f_1 = cv2.imread("video_frames/0000001670.jpg")
f_2 = cv2.imread("video_frames/0000001824.jpg")
background = np.zeros_like(f_1)

```
